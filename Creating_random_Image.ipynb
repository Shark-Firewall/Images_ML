```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating random 2d array of size 256X256 with value in btw 0 to 256
random_image=np.random.randint(0,256,(256,256))
```

```python
# Intialize plot
plt.figure(figsize=(7,7))
```

```python
#showing Grey scale image
plt.imshow(random_image,cmap="gray",vmin=0,vmax=255)
plt.show()
```
